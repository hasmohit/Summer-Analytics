```python
from pydataset import data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_table('Insurance.txt', sep='\t')
data.head()
```

```python
X_train, X_test, y_train, y_test= train_test_split(data.age.values.reshape(-1,1), data.bought_insurance.values, train_size=0.9)

model=LogisticRegression()
model.fit(X_train, y_train)
```

```python
predictions=model.predict(X_test)
```

```python
predictions
```

```python
model.score(X_test, y_test)
```

```python
plt.figure(num=1)
plt.scatter(X_test, y_test, color='k', label='Actual Data')
plt.plot(X_test, predictions, 'r-', label='Logistic Regression')
plt.plot
plt.legend()
plt.title('Logistic Regression Model')
plt.show()
```
